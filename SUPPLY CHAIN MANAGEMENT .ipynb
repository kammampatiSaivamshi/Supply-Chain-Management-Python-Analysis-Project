## 1.  Inventory Levels:



```python
data['Incoming Inventory'] = data['Order quantities'] * (data['Lead times'] / data['Manufacturing lead time'])
```

```python
data['Defective Inventory'] = data['Stock levels'] * data['Defect rates']
```

```python
# Calculate Current Inventory
data['Current Inventory'] = (data['Stock levels'] +
                           data['Incoming Inventory'] -
                           data['Number of products sold'] -
                           data['Defective Inventory'])
```

```python
# Calculate Inventory Costs
data['Inventory Costs'] = (
    data['Stock levels'] * data['Price'] +
    data['Shipping costs'] +
    data['Costs'] +
    data['Manufacturing costs'])
```

### Inventory Trend Overtime:

```python
# Create a sequential index to represent time
data['Time Period'] = range(1, len(data) + 1)

# Calculate Current Inventory
data['Current Inventory'] = (
    data['Stock levels'] + data['Incoming Inventory'] -
    data['Number of products sold'] - data['Defective Inventory']
)

# Group by Time Period to see trends
trends = data.groupby('Time Period').agg(
    Total_Current_Inventory=('Current Inventory', 'sum'),
    Total_Number_Sold=('Number of products sold', 'sum'),
    Total_Incoming_Inventory=('Incoming Inventory', 'sum'),
    Total_Defective_Inventory=('Defective Inventory', 'sum')
).reset_index()

# Plot Inventory Trends
plt.figure(figsize=(12, 6))
plt.plot(trends['Time Period'], trends['Total_Current_Inventory'], marker='o', label='Total Current Inventory')
plt.plot(trends['Time Period'], trends['Total_Number_Sold'], marker='o', label='Total Number Sold')
plt.plot(trends['Time Period'], trends['Total_Incoming_Inventory'], marker='o', label='Total Incoming Inventory')
plt.plot(trends['Time Period'], trends['Total_Defective_Inventory'], marker='o', label='Total Defective Inventory')

plt.title('Inventory Trends Over Time')
plt.xlabel('Time Period')
plt.ylabel('Inventory Metrics')
plt.xticks(trends['Time Period'])
plt.legend()
plt.grid()
plt.show()
```

- There are significant negative spikes in the total current inventory, suggesting periods of high demand where the available stock dips drastically.
- Sales show extreme peaks at various intervals, indicating that demand for products is highly variable. There are periods of high sales followed by lower periods.
- The incoming inventory seems to be relatively stable but fails to react dynamically to the drastic changes in current inventory and sales, causing mismatches between supply and demand.
- There is a moderate presence of defective inventory throughout the time period, though not excessive. However, spikes in defective inventory can affect overall supply chain performance, potentially leading to order fulfillment delays and additional costs.

## 2. Supplier Performance Analysis:
- Metric: Evaluate the performance of different suppliers based on lead times, defect rates, and on-time delivery.
- Objective: Identify the best and worst-performing suppliers.

```python
# Calculate Total Delivery Time
data['Total Delivery Time'] = data['Lead time'] + data['Shipping times']

# Calculate Average Delivery Time by Supplier
average_delivery_time_by_supplier = data.groupby('Supplier name')['Total Delivery Time'].mean().reset_index()

# Rename the column for clarity
average_delivery_time_by_supplier.rename(columns={'Total Delivery Time': 'Average Delivery Time'}, inplace=True)
print(average_delivery_time_by_supplier)
```

### Results:

- **Supplier 1 (20.85 days)** and **Supplier 4 (20.78 days)** have very close delivery times. Both suppliers could be considered for urgent orders due to their faster delivery times.
- **Supplier 2 (24.05 days)** and **Supplier 5 (24.28 days)** have similar average delivery times, falling in the middle range. While not the fastest, their consistency could be leveraged for orders that do not require urgent fulfillment.
- If reducing lead times is a priority, **Supplier 4 and Supplier 1** should be considered for critical operations.
- This suggests that Supplier 3 may be less efficient or operates with longer lead times compared to the other suppliers (require closer management, negotiation with suppliers)

```python
# Group by Time Period and calculate total Incoming and Defective Inventory
grouped_df = data.groupby('Time Period').agg({
    'Incoming Inventory': 'sum',
    'Defective Inventory': 'sum'
}).reset_index()

# Calculate Defect Rate
grouped_df['Defect Rates'] = (grouped_df['Defective Inventory'] / grouped_df['Incoming Inventory']) * 100

# Output the defect rates over time
print("Defect Rates Over Time:")
print(grouped_df[['Time Period', 'Defect Rates']])

# Example visualization (optional)

plt.plot(grouped_df['Time Period'], grouped_df['Defect Rates'], marker='o')
plt.xlabel('Time Period')
plt.ylabel('Defect Rate (%)')
plt.title('Defect Rate Over Time')
plt.xticks(rotation=90)
plt.show()
```

- The spikes indicate periods of higher defect rates, potentially due to poor supplier performance or quality control issues. Regular monitoring and evaluation of suppliers should be implemented to address the variability in defect rates.
- High defect rates can disrupt inventory levels
- Improve quality assurance processes before shipping to reduce delays in order fulfillment.
- Periods of high defects are likely increasing overall supply chain costs due to rework, returns, and additional shipping costs.

```python
# Calculate Total Orders
total_orders = data['Order quantities'].sum()

# Calculate On-Time Orders
data['On-Time'] = data['Shipping times'] <= data['Lead times']
on_time_orders = data[data['On-Time']]['Order quantities'].sum()

# Calculate On-Time Fulfillment Rate
on_time_fulfillment_rate = (on_time_orders / total_orders) * 100

# Print results
print(f'Total Orders: {total_orders}')
print(f'On-Time Orders: {on_time_orders}')
print(f'On-Time Fulfillment Rate: {on_time_fulfillment_rate:.2f}%')

# Plotting
categories = ['Total Orders', 'On-Time Orders']
values = [total_orders, on_time_orders]

plt.figure(figsize=(5, 5))
plt.bar(categories, values, color=['skyblue', 'lightgreen'])
plt.title('Order Fulfillment Overview \n')
plt.ylabel('Number of Orders')
plt.ylim(0, max(values) + 50)  # Set y-axis limit for better visibility
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding value annotations
for i, value in enumerate(values):
    plt.text(i, value + 5, str(value), ha='center', va='bottom')

plt.show()
```

### Results:
- The business has a strong on-time fulfillment rate of over **85%**, which is a **positive indicator of efficiency and customer satisfaction.**
- However, there is room for improvement, as about **14.69%** of orders are not delivered on time, which could lead to customer dissatisfaction or lost opportunities for repeat business.
- Focusing on **reducing delays and improving delivery processes** could help boost the on-time fulfillment rate and enhance overall operational performance.
